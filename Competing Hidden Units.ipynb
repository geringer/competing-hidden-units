```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt

mat = scipy.io.loadmat('mnist_all.mat')

classes = 10 
in_dim = 784 
samples = 60000 
m = np.zeros((0,in_dim))
for i in range(classes):
    m = np.concatenate((m, mat['train'+str(i)]), axis=0)
    
m=m/255.0
```

```python
def draw_weights(synapses, Kx, Ky):
    yy=0
    HM=np.zeros((28*Ky,28*Kx))
    for y in range(Ky):
        for x in range(Kx):
            HM[y*28:(y+1)*28,x*28:(x+1)*28]=synapses[yy,:].reshape(28,28)
            yy += 1
    plt.clf()
    nc=np.amax(np.absolute(HM))
    im=plt.imshow(HM,cmap='bwr',vmin=-nc,vmax=nc)
    fig.colorbar(im,ticks=[np.amin(HM), 0, np.amax(HM)])
    plt.axis('off')
    fig.canvas.draw()
    
def draw_neuron(synapses, row = 0, random = False):
    if random:
        choices = synapses.shape[0]
        row = np.random.randint(choices)
        
    if synapses == "Tesnor":
        synapses = synapses.cpu().detach().numpy()
        
    HM = synapses[row, :]
    plt.clf()
    nc=np.amax(np.absolute(HM))
    im=plt.imshow(HM.reshape((28, 28)),cmap='bwr',vmin=-nc,vmax=nc)
    plt.axis('off')
    plt.title(str(row) + "th Neuron")
    fig.subplots_adjust(left = 0) 
    fig.colorbar(im, ticks=[np.amin(HM), 0, np.amax(HM)])
    fig.canvas.draw()
    plt.show()
    
fig = plt.figure(figsize=(12.9,10))
device = "cuda"
eps0 = 2e-2    # initial learning rate
Kx = 15
Ky = 15
hid = 1000#Kx*Ky    # number of hidden units
mu = 0.0       # mean for initialization
sigma = 1.0    # SD for initialization
epochs = 200   # number of epochs
minibatch_size = 100      # size of the minibatch
prec = 1e-30   # Machine eps
delta = 0.4    # Strength of the anti-hebbian learning
p = 2.0        # Lebesgue norm of the weights
k = 2          # ranking parameter, must be integer that is bigger or equal than 2
```

```python
import torch.distributions as tdist
import torch

M = torch.from_numpy(m).float().to(device)
tsynapses = tdist.Normal(mu, sigma).sample((hid, in_dim)).to(device)
# tsynapses = torch.from_numpy(synapses).float().to(device)

for epoch in range(epochs):
    print("\n================\n", epoch, "th epoch, epsilon = ", eps)
    draw_neuron(tsynapses.cpu().detach().numpy())
    
    eps = eps0 * ( 1 - epoch / epochs)
    teps = torch.tensor([eps]).to(device)
    M = M[torch.randperm(samples),:]
    
    for i in range(samples // minibatch_size):
        
        #Take the ith row, make it a column
        inputs = M[i * minibatch_size:(i + 1) * minibatch_size, :].t()

        #take the sign of each synapse 
        sig = torch.sign(tsynapses)
        
        #dot product of the synapses * inputs 
        tot_input = torch.mm((sig * torch.abs(tsynapses) ** (p-1)), inputs)

        #returns matrix of same size with each column ranked 0-99
        y = torch.argsort(tot_input, dim=0)
        
        #creates new matrix of same shape to populate
        yl = torch.zeros((hid, minibatch_size)).float().to(device)

        #populating the hid columns and the number of the batch size as 1
        yl[y[hid-1,:], torch.arange(minibatch_size)] = 1.0
        yl[y[hid-k], torch.arange(minibatch_size)] = -delta

        #infinity norm(sum of rows) of theelementwise multiplication
        xx = torch.sum(yl * tot_input, dim = 1)
        
        #dot product of new matrix with the inputs
        A = torch.mm(yl, inputs.t())
        
        C = xx.reshape(xx.size()[0], 1).repeat(1, in_dim) #tile
        
        #matrix mutliplication of synapses with tiled xx
        B = torch.mul(C, tsynapses)
    
        ds =  A - B

        #max absolute of DS
        nc = torch.max(torch.abs(ds))
        
        #Change this for float32
        prec = torch.tensor([prec]).float().to(device)
        if nc < prec:
            nc = prec
            
        #adds a small value to the synapses matrix
        tsynapses += teps * torch.div(ds, nc)
            
```

```python
import torch.nn as nn
import torch.nn.functional as F
import torch.utils.data.dataloader as dataloader
import torch.optim as optim

from torch.autograd import Variable
from torchvision import transforms

class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.fc = nn.Linear(784, 1000)
        self.fc2 = nn.Linear(1000, 10)

    def forward(self, x):
        x = x.view((-1, 784))
        h = F.relu(self.fc(x))
        h = self.fc2(h)
        return F.log_softmax(h)    
    
    
model = Model().cuda()
model.fc.weight.data = tsynapses

optimizer = optim.Adam(model.parameters(), lr=1e-3)
criterion = nn.CrossEntropyLoss()

count = 0
for w in model.parameters():
    if count == 0:
        w.requires_grad = False
        count += 1
    try:
        plt.imshow(w.cpu().detach().numpy(), interpolation='nearest')
        plt.show()
    except:
        pass
    finally:
        print(w.size(), w.requires_grad, "\n")
```
